```python

import pandas as pd
data = pd.read_csv("speed-profiling.csv", index_col=0)
```

```python
import seaborn as sns
sns.set_context("talk")
fig, ax = plt.subplots(figsize=(5, 5))
ax = sns.swarmplot(x="Sequences", y="Time", hue="Implementation", data=data) 
ax.set_yscale("log")
ax.set_xlabel("Number of Sequences")
ax.set_ylabel("Time (seconds)")
sns.despine()
plt.tight_layout()
plt.savefig("speed-comparison.png", dpi=300)
```
